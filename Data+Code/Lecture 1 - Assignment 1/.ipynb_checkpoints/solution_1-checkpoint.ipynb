# A01.1 Building and visualising networks

Parameters: Build each network with ð‘ = 32 nodes. Note
that these networks are undirected and unweighted.

## Star network
For ð‘ vertices: ð‘˜1 = ð‘ âˆ’ 1; ð‘˜ð‘– = 1, for ð‘– > 1

```python
g2 = nx.Graph()
g2.add_node(100)
for i in range(32):
    g2.add_node(i)
    g2.add_edge(100,i)
nx.draw(g2, with_labels = False)
```

## Fully-Connected

For ð‘ vertices: ð‘˜ð‘– = ð‘ âˆ’ 1

```python
g2 = nx.Graph()
for i in range(32):
    g2.add_node(i)
    for j in range(g2.number_of_nodes()) :
        if (j!=i) :
            g2.add_edge(i,j)
nx.draw(g2, with_labels = False)
```

```python
g2_degrees = g2.degree()
degrees = [ g2_degrees[i] for i in g2.nodes() ]
print("Edg", g2.number_of_edges())
print("Nod", g2.number_of_nodes())
print("Min", min(degrees))
print("Max", min(degrees))
```

## One-dimensional lattice with periodic boundary conditions
where each node ð‘– has degree ð‘˜ð‘– = 2


```python
g2 = nx.Graph()
for i in range(32):
    g2.add_node(i)
    if (g2.number_of_nodes()>1) :
        g2.add_edge(i,i-1)
        
g2.add_edge(0,31)
nx.draw(g2, with_labels = False)
```

```python
g2_degrees = g2.degree()
degrees = [ g2_degrees[i] for i in g2.nodes() ]
print("Edg", g2.number_of_edges())
print("Nod", g2.number_of_nodes())
print("Min", min(degrees))
print("Max", min(degrees))
```
