```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
def myfunc(x,alpha):
    return x**(-alpha)
```

```python
N,m=100,1
g = nx.barabasi_albert_graph(N,m)
```

```python
nx.draw(g)
```

```python
# We create an array with all degrees
k = np.asarray([d for d in dict(g.degree()).values()])
```

```python
k)
```

```python
spacing = np.linspace(min(k),max(k),len(k))
spacing
```

```python
#Install the powerlaw package
pip install powerlaw
```

```python
import powerlaw
fit = powerlaw.Fit(k)
```

```python
alpha = fit.power_law.alpha
```

```python
alpha
```

```python
sigma = fit.power_law.sigma
```

```python
sigma
```

```python
fit.power_law.plot_pdf(color='r')
powerlaw.plot_pdf(k,color='b',marker='o')
plt.plot(spacing,myfunc(spacing,4))
plt.show()
```
